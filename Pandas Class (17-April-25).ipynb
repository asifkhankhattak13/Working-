```python
import pandas as pd

data = {
    "Age group": ["18-25", "25-35", "35-45", "45-55", "55-65", "65+", "Total"],
    "Employment %": ["60%", "85%", "95%", "97%", "97%", "100%", "—"],
    "Employed": [60, 85, 95, 97, 97, 100, 534],
    "Unemployed": [40, 15, 5, 3, 3, 0, 66],
    "Total employees": [100, 100, 100, 100, 100, 100, 600]
}

df = pd.DataFrame(data)
df_plot = df[df["Age group"] != "Total"]
print(df)
```

```python
import matplotlib.pyplot as plt
```

```python
x = df_plot["Age group"]
width = 0.35  # width of the bars

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(x, df_plot["Employed"], width, label='Employed', color='green')
ax.bar(x, df_plot["Unemployed"], width, bottom=df_plot["Employed"], label='Unemployed', color='blue')

# Labels and legend
ax.set_title('Employment by Age Group')
ax.set_xlabel('Age Group')
ax.set_ylabel('Number of People')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np

x = np.arange(len(df_plot["Age group"]))
width = 0.35  # Width of each bar

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, df_plot["Employed"], width, label='Employed', color='teal')
bars2 = ax.bar(x + width/2, df_plot["Unemployed"], width, label='Unemployed', color='gray')

# Labels and formatting
ax.set_title('Employed vs Unemployed by Age Group')
ax.set_xlabel('Age Group')
ax.set_ylabel('Number of People')
ax.set_xticks(x)
ax.set_xticklabels(df_plot["Age group"])
ax.legend()
plt.tight_layout()
plt.show()
```
